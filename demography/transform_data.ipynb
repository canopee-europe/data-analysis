```python
import pandas as pd
```

```python
df = pd.read_csv('pop_par_sexe_et_age.csv')
df
```

```python
df_long = df.melt(id_vars=["Année", "Population au premier janvier"],
                  var_name="Sexe_Tranche",
                  value_name="Population")

df_long
```

```python
df_long[['Sexe', 'Tranche d\'âge']] = df_long['Sexe_Tranche'].str.extract(r'(Homme|Femme) - (.+)')
```

```python
df_long.drop(columns=["Sexe_Tranche"], inplace=True)
```

```python
df_total = df[["Année", "Population au premier janvier"]].copy()
df_total["Sexe"] = "Tout"
df_total["Tranche d'âge"] = "Tout"
df_total.rename(columns={"Population au premier janvier": "Population"}, inplace=True)
```

```python
df_final = pd.concat([df_total, df_long], ignore_index=True)
df_final
```

```python
df_final.to_csv("population_long.csv", index=False)
```
