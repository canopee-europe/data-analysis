```python
import requests
import pandas as pd
import csv
import numpy as np
```

```python
cleaned_rows = []

with open('geo_prog_2019.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        cleaned_rows.append([col for col in row if col.strip()])
```

```python
df = pd.DataFrame(cleaned_rows)
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df.columns = ["Année", "Région", "Type de mesure", "Nationalité", "Mesure"]
df
```
